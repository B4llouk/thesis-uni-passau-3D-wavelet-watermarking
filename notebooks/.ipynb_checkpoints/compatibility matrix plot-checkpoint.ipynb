```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Define the data for the compatibility matrix
data = {
    'Feature 1': [1, 0, 1],
    'Feature 2': [0, 1, 0]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data, index=['Item 1', 'Item 2', 'Item 3'])

# Create a figure and axis for the plot
plt.figure(figsize=(8, 6))

# Create a heatmap
sns.heatmap(df, annot=True, cmap="YlGnBu", cbar=False, square=True)

# Add labels and title
plt.title('Compatibility Matrix')
plt.ylabel('Items')
plt.xlabel('Features')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the data for the compatibility matrix
data = {
    'Feature 1': [1, 0, 1],
    'Feature 2': [0, 1, 0]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data, index=['Item 1', 'Item 2', 'Item 3'])

# Create a figure and axis for the plot
plt.figure(figsize=(8, 6))

# Create a heatmap
sns.heatmap(df, annot=True, cmap="YlGnBu", cbar=False, square=True, yticklabels=True)

# Add labels and title
plt.title('Compatibility Matrix')
plt.xlabel('Features')

# Adjust the position of the y-axis labels
plt.yticks(rotation=0) 

# Show the plot
plt.show()
```
